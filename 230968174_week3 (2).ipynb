```python
# Experiment 1: Smart Thermostat Agent (Simple Reflex Agent)

class SmartThermostatAgent:
    def __init__(self):
        self.states = ["Low", "Comfortable", "High"]

    def perceive_and_act(self, temperature):
        # Condition-Action Rules
        if temperature == "Low":
            return "Heat"
        elif temperature == "High":
            return "Cool"
        else:
            return "Idle"

    def update_temperature(self, temperature, action):
        index = self.states.index(temperature)

        if action == "Heat" and index < 2:
            return self.states[index + 1]
        elif action == "Cool" and index > 0:
            return self.states[index - 1]
        else:
            return temperature


# Simulation
agent = SmartThermostatAgent()
initial_states = ["Low", "Comfortable", "High"]

print("Smart Thermostat Simulation:\n")

for state in initial_states:
    print(f"Initial Temperature: {state}")
    temp = state
    for step in range(5):
        action = agent.perceive_and_act(temp)
        temp = agent.update_temperature(temp, action)
        print(f"Step {step+1}: Action = {action}, Temperature = {temp}")
    print("-" * 40)
```

```python
# Experiment 2: Elevator Control Agent (Goal-Based Agent)

class ElevatorAgent:
    def __init__(self):
        self.current_floor = 1

    def move(self, request_floor):
        print(f"\nRequest received for Floor {request_floor}")

        while self.current_floor != request_floor:
            if self.current_floor < request_floor:
                self.current_floor += 1
                print("Moving Up → Floor", self.current_floor)
            elif self.current_floor > request_floor:
                self.current_floor -= 1
                print("Moving Down → Floor", self.current_floor)

        print("Reached requested floor:", self.current_floor)


# Simulation
elevator = ElevatorAgent()
requests = [3, 1, 2]

for req in requests:
    elevator.move(req)
```

```python
# Experiment 3: Autonomous Traffic Signal Agent (Utility-Based Agent)

class TrafficSignalAgent:
    def __init__(self):
        # Utility values: higher means more priority
        self.utility = {
            "Low": 1,
            "High": 5
        }

    def decide_signal(self, road_a, road_b):
        utility_a = self.utility[road_a]
        utility_b = self.utility[road_b]

        if utility_a > utility_b:
            return "Green for Road A"
        else:
            return "Green for Road B"


# Simulation
agent = TrafficSignalAgent()

traffic_scenarios = [
    ("High", "Low"),
    ("Low", "High"),
    ("High", "High"),
    ("Low", "Low")
]

print("Traffic Signal Simulation:\n")

for road_a, road_b in traffic_scenarios:
    decision = agent.decide_signal(road_a, road_b)
    print(f"Road A: {road_a}, Road B: {road_b} → Signal: {decision}")
```

```python
# Experiment 4: Warehouse Robot Navigation Agent (Model-Based Agent)

class WarehouseRobot:
    def __init__(self, grid, start, target):
        self.grid = grid
        self.position = start
        self.target = target
        self.visited = set()

    def move(self, direction):
        x, y = self.position

        if direction == "Up":
            x -= 1
        elif direction == "Down":
            x += 1
        elif direction == "Left":
            y -= 1
        elif direction == "Right":
            y += 1

        if 0 <= x < len(self.grid) and 0 <= y < len(self.grid[0]):
            if self.grid[x][y] == 0:  # 0 means free cell
                self.position = (x, y)
                self.visited.add(self.position)
                return True
        return False

    def navigate(self):
        print("Starting Navigation...\n")

        while self.position != self.target:
            x, y = self.position

            # Try moving right then down (simple strategy)
            if y < self.target[1] and self.move("Right"):
                pass
            elif x < self.target[0] and self.move("Down"):
                pass
            else:
                break

            print("Robot at position:", self.position)

        if self.position == self.target:
            print("\nTarget reached at:", self.target)
        else:
            print("\nFailed to reach target.")


# Grid: 0 = free, 1 = obstacle
warehouse = [
    [0, 0, 0],
    [1, 0, 0],
    [0, 0, 0]
]

start = (0, 0)
target = (2, 2)

robot = WarehouseRobot(warehouse, start, target)
robot.navigate()
```

```python
# Experiment 5: Online Recommendation Agent (Learning Agent)

import random

class RecommendationAgent:
    def __init__(self):
        self.product_scores = {
            "A": 0,
            "B": 0
        }

    # Performance Element
    def recommend(self):
        if self.product_scores["A"] >= self.product_scores["B"]:
            return "A"
        else:
            return "B"

    # Learning Element
    def learn(self, product, feedback):
        if feedback == "like":
            self.product_scores[product] += 1
        else:
            self.product_scores[product] -= 1

    # Critic
    def get_feedback(self):
        return random.choice(["like", "dislike"])

    # Problem Generator
    def explore(self):
        return random.choice(["A", "B"])


# Simulation
agent = RecommendationAgent()

print("Recommendation System Simulation:\n")

for session in range(10):
    if random.random() < 0.7:
        product = agent.recommend()
    else:
        product = agent.explore()

    feedback = agent.get_feedback()
    agent.learn(product, feedback)

    print(f"Session {session+1}: Recommended Product {product} → User Feedback: {feedback}")
    print("Current Scores:", agent.product_scores)
    print("-" * 30)
```
