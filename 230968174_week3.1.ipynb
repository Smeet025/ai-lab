```python
from collections import deque
import time
```

EXPERIMENT 1: BFS – Maze Navigation

```python
maze = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E', 'G'],
    'G': []
}

start_node = 'A'
goal_node = 'G'
# Maze is represented as a graph using adjacency list with a start and goal node
```

```python
def bfs(graph, start, goal):
    queue = deque([[start]])
    visited = set([start])
    nodes_expanded = 0

    while queue:
        path = queue.popleft()
        node = path[-1]
        nodes_expanded += 1

        if node == goal:
            return path, nodes_expanded

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])

    return None, nodes_expanded
# This function performs Breadth First Search and finds the shortest path in the maze
```

```python
path, expanded = bfs(maze, start_node, goal_node)
print("BFS Path:", path)
print("Nodes Expanded:", expanded)
# BFS is executed here and the resulting path and number of expanded nodes are printed
```

EXPERIMENT 2: DFS – Graph Traversal

```python
def dfs(graph, current, goal, path, visited):
    visited.add(current)
    path.append(current)

    if current == goal:
        return path

    for neighbor in graph[current]:
        if neighbor not in visited:
            result = dfs(graph, neighbor, goal, path, visited)
            if result:
                return result

    path.pop()
    return None
# This function implements Depth First Search using recursion and backtracking

```

```python
dfs_path = dfs(maze, start_node, goal_node, [], set())
print("DFS Path:", dfs_path)
# DFS is called here to find a path from start node to goal node
```

EXPERIMENT 3: 8-Puzzle Using BFS

```python
initial_state = (1, 2, 3,
                 4, 0, 5,
                 6, 7, 8)

goal_state = (1, 2, 3,
              4, 5, 6,
              7, 8, 0)
# Initial and goal configurations of the 8-puzzle are defined using tuples
```

```python
def get_neighbors(state):
    neighbors = []
    idx = state.index(0)

    moves = [-3, 3, -1, 1]

    for move in moves:
        new_idx = idx + move
        if 0 <= new_idx < 9:
            new_state = list(state)
            new_state[idx], new_state[new_idx] = new_state[new_idx], new_state[idx]
            neighbors.append(tuple(new_state))

    return neighbors
# This function generates all possible valid moves of the blank tile
```

```python
def bfs_8_puzzle(start, goal):
    queue = deque([[start]])
    visited = set([start])
    expanded = 0

    while queue:
        path = queue.popleft()
        state = path[-1]
        expanded += 1

        if state == goal:
            return path, expanded

        for neighbor in get_neighbors(state):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])

    return None, expanded
# BFS is used here to find the shortest solution path for the 8-puzzle problem
```

```python
solution, expanded_states = bfs_8_puzzle(initial_state, goal_state)

print("Solution Path:")
for step in solution:
    print(step)

print("Total States Expanded:", expanded_states)
# The solution steps and total number of expanded states are displayed here
```

EXPERIMENT 4: BFS vs DFS Performance Comparison

```python
start_time = time.time()
bfs(maze, start_node, goal_node)
bfs_time = time.time() - start_time
# Execution time for BFS is calculated using the time module
```

```python
start_time = time.time()
dfs(maze, start_node, goal_node, [], set())
dfs_time = time.time() - start_time
# Execution time for DFS is calculated to compare with BFS
```

```python
print("Algorithm | Time Taken (seconds)")
print("-------------------------------")
print(f"BFS       | {bfs_time:.6f}")
print(f"DFS       | {dfs_time:.6f}")
# Time taken by BFS and DFS is displayed in a simple comparison format
```

1. BFS gives optimal solution because it explores nodes level by level.
2. DFS is useful when memory is limited and solution depth is unknown.
3. Higher branching factor increases the time and memory required for BFS.
4. BFS is not suitable for large state spaces because it needs a lot of memory.
